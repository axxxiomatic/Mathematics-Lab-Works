```python
#  1
import numpy as np

R = np.array( [ [1, 2, 3, 4], [2, 3, 4, 5], [3, 4, 5, 6] ] )
print("R:\n", R, "\n")

#  2
R[2, 3] = -R[2, 3]
R[0, 2] = R[0, 2] - 4
print("R:\n", R, "\n")

R[:, 1] = R[:, 1] * 2
R[0, :] = R[0, :] * 3
print("R:\n", R, "\n")

#  3
A = np.arange(1, 5, 1)
B = np.arange(1, 7, 1)
print("A:\n", A, "\n\nB:\n", B, "\n")

C = np.arange(8).reshape(8, 1)
D = np.arange(10).reshape(10, 1)
E = np.arange(3).reshape(3, 1)
print("C:\n", C, "\n\nD:\n", D, "\n\nE:\n", E)
```

```python
import numpy as np
import numpy.matlib

#  1
A = np.array( [ [1, 2, 3], [4, 5, 6] ] )
B = np.array( [ [1, -2, 1], [-2, 3, 4] ] )
c = 2
D = np.matlib.ones((2, 3))
E = np.matlib.eye((3))
print("A:\n", A, "\n\nB:\n", B, "\n\nc = ", c, "\n\nD:\n", D, "\n\nE:\n", E, "\n")

#  2
print("A + B =\n", A + B, "\nA + c =\n", A + c, "\n")

try:
    print(A + E)
except Exception:
    print('     .')
    
print("\nA - B =\n", A - B, "\n\nA - c =\n", A - c, "\n\nc * A =\n", c * A, "\n\nA * B =\n", A * B, "\n")

try:
    print(A * B.transpose())
except Exception:
    print('       .')
```
